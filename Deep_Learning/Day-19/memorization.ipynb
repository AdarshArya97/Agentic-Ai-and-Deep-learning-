## This is exponential approach it will take more time for bigger numbers


```python
import time
def fib(n):
    if n==1 or n==0:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
start = time.time()
fib(40)
print(time.time()-start)
```

### using memorisation


```python
import time
def fib2(n,d):
    if n in d:
        return d[n]
    else:
        d[n] = fib2(n-1,d) + fib2(n-2,d)
        return d[n]
```

```python
start = time.time()
d = {0:1,1:1}
fib2(40,d)
print(time.time()-start)
```

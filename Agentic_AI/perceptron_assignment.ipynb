### A2. Does the decision boundary stay a hyperplane if we change the activation function?

Suppose the hard limiter in the perceptron is replaced with the smooth activation function

```
φ(v) = tanh(v / 2)
```

The activation φ(v) = tanh(v/2) is not an independent choice of function — it is a rescaled form of the logistic sigmoid σ(v) = 1/(1+e⁻ᵛ) already used earlier in the course, related by the identity:

```
tanh(v/2) = 2σ(v) − 1
```

where `v = w1x1 + w2x2 + b` is the induced local field, same as before. The classification rule is now:

```
assign x to class c1 if φ(v) > λ         (λ is some fixed threshold, e.g. λ = 0)
assign x to class c2 otherwise
```

**Question:** Show that the decision boundary produced by this rule is *still* a hyperplane (i.e., still of the form `v = constant` — a flat boundary, not a curved one), even though `φ` itself is a curved, nonlinear function. (Hint: `tanh` is a strictly increasing function — think about what `φ(v) > λ` implies about `v` itself.)

### My Answer:

The decision boundary is still a hyperplane because φ(v) = tanh(v/2) is a strictly increasing function. This means that if φ(v1) > φ(v2), then it must be that v1 > v2.

The classification rule is assign x to class c1 if φ(v) > λ and assign x to class c2 otherwise.

Since φ(v) = tanh(v/2) is strictly increasing, the condition φ(v) > λ is equivalent to tanh(v/2) > λ. Applying the inverse tanh function (which is also strictly increasing) to both sides, we get:

v/2 > tanh⁻¹(λ)

Let C = tanh⁻¹(λ). Since λ is a fixed threshold, C is also a constant. Therefore, the inequality becomes:

v/2 > C

or

v > 2C

Since v = w1x1 + w2x2 + b, the condition v > 2C means w1x1 + w2x2 + b > 2C. This can be rewritten as w1x1 + w2x2 + (b - 2C) > 0.

This is still in the form of w1x1 + w2x2 + b' > 0 (where b' = b - 2C), which defines a linear inequality. The decision boundary itself is given by w1x1 + w2x2 + b' = 0, which is the equation of a straight line (a hyperplane in 2D input space). Thus, the decision boundary remains a hyperplane, even with the nonlinear tanh activation function, because the thresholding operation (φ(v) > λ) ultimately translates back to a linear condition on v.

 By Replacing the hard limiter with tanh (or any strictly monotonic squashing function) warps the output values between −1 and 1 in a nonlinear, S-shaped way — but it does not reorder which side of any given hyperplane a point falls on. Since classification only cares about comparing φ(v) to a threshold, and monotonicity preserves the ordering of v, the nonlinear "bending" of φ never translates into a bending of the decision boundary itself. The curvature lives entirely in how confidently points are classified (i.e., how close φ(v) is to ±1), not in where the boundary sits.

### B1. Implement the perceptron's forward pass (decision rule only)

This is exactly the forward pass you already know — weighted sum, then a hard-limiter activation. There is no `fit()` method here on purpose.

````python
class Perceptron:
    def __init__(self, w1, w2, b):
        """Weights are supplied directly — not learned."""
        self.w1 = w1
        self.w2 = w2
        self.b = b

    def predict(self, x1, x2):
        """
        Return +1 or -1 using the sign of the induced local field v = w1*x1 + w2*x2 + b.
        """

        v = self.w1 * x1 + self.w2 * x2 + self.b
        return +1 if v > 0 else -1


    def accuracy(self, X, d):
        """
        X: array of shape (n_samples, 2). d: array of desired labels (+1/-1).
        Returns the fraction of samples correctly classified.
        """
        preds = np.array([self.predict(x1, x2) for x1, x2 in X])
        return np.mean(preds == d)
````

### B2. Plug in your hand-designed weights from A1

Use the exact `w1`, `w2`, `b` values you derived in A1. If they're correct, accuracy should be exactly `1.0` for both gates.

````python
X_and = np.array([[-1,-1],[-1,1],[1,-1],[1,1]])
d_and = np.array([-1,-1,-1, 1])

X_or  = np.array([[-1,-1],[-1,1],[1,-1],[1,1]])
d_or  = np.array([-1, 1, 1, 1])


and_gate = Perceptron(w1=0.5, w2=1.0, b=-0.5)
or_gate  = Perceptron(w1=0.5, w2=0.5, b=0.5)

print("AND accuracy:", and_gate.accuracy(X_and, d_and))
print("OR  accuracy:", or_gate.accuracy(X_or, d_or))

assert and_gate.accuracy(X_and, d_and) == 1.0, "AND weights are not correct yet"
assert or_gate.accuracy(X_or, d_or) == 1.0, "OR weights are not correct yet"
print("Both gates correctly implemented with hand-designed weights.")
````

Since no learning algorithm is available yet, we'll demonstrate the impossibility a different way: **scan a fine grid of possible weight values** and check whether *any* of them gets all 4 XOR points right. If Part A3's reasoning is correct, none should.

````python
from itertools import product
X_xor = np.array([[-1,-1],[-1,1],[1,-1],[1,1]])
d_xor = np.array([-1, 1, 1,-1])


w_range = np.arange(-1, 1.5, 0.5)

best_accuracy = 0.0
best_weights = None



for w1, w2, b in product(w_range, repeat=3):
    p = Perceptron(w1, w2, b)
    acc = p.accuracy(X_xor, d_xor)

    if acc > best_accuracy:
        best_accuracy = acc
        best_weights = (w1, w2, b)

print(f"Best accuracy found across the entire grid: {best_accuracy:.2f}")
print(f"Achieved by weights: {best_weights}")
print(f"Perfect (1.0) accuracy found: {best_accuracy == 1.0}")
````

```python
def attention_score(word_a, word_b):
    """
    Measures how much word_a should care about word_b.
    Higher score = stronger attention.
    """
    vec_a = word_vectors[word_a]
    vec_b = word_vectors[word_b]
    return sum(a * b for a, b in zip(vec_a, vec_b))
```

```python
def self_attention(word, sentence):
    """
    Each word looks at every other word
    and collects useful information.
    """

    scores = {}
    for other_word in sentence:
        scores[other_word] = attention_score(word, other_word)

    total_score = sum(scores.values())

    # Normalize scores so they behave like probabilities
    attention_weights = {
        w: scores[w] / total_score if total_score != 0 else 0
        for w in scores
    }

    return attention_weights
```

```python
def update_vector(word, sentence):
    """
    The word updates its knowledge
    based on what it learned from others.
    """

    weights = self_attention(word, sentence)
    new_vector = [0, 0, 0]

    for other_word, weight in weights.items():
        for i in range(len(new_vector)):
            new_vector[i] += weight * word_vectors[other_word][i]

    return new_vector

    updated_vectors = {}
    for word in sentence:
        updated_vectors[word] = update_vector(word, sentence)

    print("\nUpdated word knowledge after discussion:")
    for word in sentence:
        print(word, "→", [round(v, 2) for v in updated_vectors[word]])



    
```

```python
def feed_forward(vector):

    """

    A simple processing step applied independently

    to each word.

    """

    return [v * 1.5 for v in vector]

final_vectors = {
    word: feed_forward(vec)
    for word, vec in update_vector.items()
}

print("\nFinal word knowledge after individual thinking:")
for word in sentence:
    print(word,"->",[round(v,2) for v in final_vectors])
```

```python
# Pure Python: Building a "Baby" Tokenizer
# No libraries. Just logic.

vocab = {
    "hello": 1,
    "world": 2,
    "!": 3,
    "is": 4,
    "fun": 5
}

print("Vocabulary (Word -> ID):")
for word, idx in vocab.items():
    print(f"{word} -> {idx}")


# Step 1: Text → Token IDs (Encoding)
# --------------------------------------------------------

def encrypt(text):
    """
    Converts human-readable text into numbers.
    This is what tokenization + encoding looks like.
    """

    print("\nOriginal text:", text)

    # Make text lowercase and separate punctuation
    words = text.lower().replace("!", " !").split()
    print("After basic splitting:", words)

    ids=[]
    for word in words:
        if word in vocab:
            ids.append(vocab[word])
        else:
            ids.append(0)
    return ids
```

```python
message = "Hello world!"
token_ids = encrypt(message)

print("\nToken IDs:",token_ids)

# decoded_text = decrypt(token_ids)
# print("Decoded text:",decoded_text)


```
